```python
# =========================================================
# GOOGLE COLAB - TGDĐ INVENTORY DATA GENERATOR
# AI DEMAND FORECASTING & STOCKOUT RISK
# RETAIL STORE LEVEL DATA
# =========================================================

# =========================================================
# 1. IMPORT LIBRARIES
# =========================================================

import pandas as pd
import numpy as np
import random

# =========================================================
# 2. RANDOM SEED
# =========================================================

np.random.seed(42)
random.seed(42)

# =========================================================
# 3. STORE IDs
# =========================================================

store_ids = [
    "TGDĐ_001",
    "TGDĐ_002",
    "TGDĐ_003"
]

# =========================================================
# 4. SKU LIST
# =========================================================

skus = [

    # ================= APPLE =================

    {
        "sku_id": "IP16-PRM-256-TITAN",
        "brand": "Apple",
        "is_hot": True
    },

    {
        "sku_id": "IP16-128-BLACK",
        "brand": "Apple",
        "is_hot": False
    },

    {
        "sku_id": "MAC-AIR-M4-8-256",
        "brand": "Apple",
        "is_hot": True
    },

    {
        "sku_id": "MAC-PRO-M4-18-512",
        "brand": "Apple",
        "is_hot": True
    },

    {
        "sku_id": "IPAD-PRO-M4-256",
        "brand": "Apple",
        "is_hot": False
    },

    # ================= SAMSUNG =================

    {
        "sku_id": "SS-S25-ULTRA-512",
        "brand": "Samsung",
        "is_hot": True
    },

    {
        "sku_id": "SS-S25-256-BLUE",
        "brand": "Samsung",
        "is_hot": False
    },

    {
        "sku_id": "SS-ZFOLD6-512",
        "brand": "Samsung",
        "is_hot": True
    },

    {
        "sku_id": "SS-ZFLIP6-256",
        "brand": "Samsung",
        "is_hot": False
    },

    {
        "sku_id": "SS-A55-128-BLK",
        "brand": "Samsung",
        "is_hot": False
    }
]

# =========================================================
# 5. DATE RANGE
# =========================================================

dates = pd.date_range(
    start="2025-01-01",
    end="2026-05-31"
)

# =========================================================
# 6. GENERATE DATA
# =========================================================

data = []

for date in dates:

    month = date.month

    # =====================================================
    # SEASONAL EVENTS
    # =====================================================

    is_summer = month in [5, 6, 7]
    is_bts = month in [8, 9]
    is_tet = month in [1, 2]
    is_year_end = month in [11, 12]

    for store in store_ids:

        for item in skus:

            sku = item["sku_id"]
            brand = item["brand"]
            is_hot = item["is_hot"]

            # =================================================
            # PRODUCT LAUNCH EVENTS
            # =================================================

            is_apple_launch = (
                brand == "Apple"
                and month in [9, 10]
            )

            is_samsung_s_launch = (
                brand == "Samsung"
                and "S24" in sku
                and month in [1, 2]
            )

            is_samsung_z_launch = (
                brand == "Samsung"
                and "ZFOLD" in sku
                and month in [7, 8]
            )

            is_new_launch = (
                is_apple_launch
                or is_samsung_s_launch
                or is_samsung_z_launch
            )

            # =================================================
            # PROMOTION FLAG
            # =================================================

            promo_prob = 0.10

            # Samsung thường sale nhiều hơn
            if brand == "Samsung":
                promo_prob += 0.10

            # Máy không hot sale nhiều hơn
            if not is_hot:
                promo_prob += 0.20

            # Mùa hè + tựu trường
            if is_summer or is_bts:
                promo_prob += 0.15

            # Apple mới ra mắt ít sale
            if is_apple_launch and brand == "Apple":
                promo_prob = 0.03

            promo_flag = np.random.choice(
                [0, 1],
                p=[1 - promo_prob, promo_prob]
            )

            # =================================================
            # SALES QUANTITY
            # =================================================

            # Apple flagship bán mạnh nhất
            if brand == "Apple" and is_hot:

                base_sales = np.random.uniform(10, 18)

            elif brand == "Apple":

                base_sales = np.random.uniform(4, 10)

            elif brand == "Samsung" and is_hot:

                base_sales = np.random.uniform(6, 14)

            else:

                base_sales = np.random.uniform(2, 7)

            multiplier = 1.0

            # Summer
            if is_summer:
                multiplier *= 1.15

            # Back to school
            if is_bts:
                multiplier *= 1.30

            # Tết
            if is_tet:
                multiplier *= 1.10

            # Year end
            if is_year_end:
                multiplier *= 1.20

            # Product launch
            if is_new_launch:
                multiplier *= 1.50

            # Flash sale
            if promo_flag == 1:
                multiplier *= 1.15

            # Random noise
            multiplier *= np.random.uniform(0.90, 1.10)

            sales_qty = int(
                round(base_sales * multiplier)
            )

            sales_qty = max(1, sales_qty)

            # =================================================
            # LEAD TIME
            # =================================================

            # Máy bán chạy được ưu tiên replenish

            if sales_qty >= 18:

                lead_time_days = random.randint(2, 5)

            elif sales_qty >= 10:

                lead_time_days = random.randint(4, 7)

            else:

                lead_time_days = random.randint(7, 14)

            # =================================================
            # CURRENT STOCK
            # =================================================

            # Retail store inventory realism

            if brand == "Apple" and is_hot:

                # iPhone Pro Max / MacBook hot
                current_stock = random.randint(3, 10)

            elif brand == "Apple":

                # iPhone thường / iPad
                current_stock = random.randint(5, 15)

            elif brand == "Samsung" and is_hot:

                # Galaxy S Ultra / Fold
                current_stock = random.randint(4, 12)

            else:

                # Samsung mid-range
                current_stock = random.randint(8, 20)

            # Sales cao -> stock giảm
            if sales_qty >= 18:

                current_stock -= random.randint(2, 5)

            elif sales_qty >= 10:

                current_stock -= random.randint(1, 3)

            current_stock = max(2, current_stock)

            # =================================================
            # FORECAST DEMAND
            # =================================================

            future_days = random.randint(7, 14)

            forecast_multiplier = np.random.uniform(0.95, 1.10)

            forecast_demand = int(
                round(
                    sales_qty
                    * future_days
                    * forecast_multiplier
                )
            )

            # =================================================
            # SAVE ROW
            # =================================================

            data.append([
                date.strftime("%Y-%m-%d"),
                store,
                sku,
                sales_qty,
                lead_time_days,
                current_stock,
                promo_flag,
                forecast_demand
            ])

# =========================================================
# 7. CREATE DATAFRAME
# =========================================================

columns = [
    "date",
    "store_id",
    "sku_id",
    "sales_qty",
    "lead_time_days",
    "current_stock",
    "promo_flag",
    "forecast_demand"
]

df = pd.DataFrame(data, columns=columns)

# =========================================================
# 8. CREATE STOCKOUT RISK
# =========================================================

# Risk score:
# Demand cao + Forecast cao + Stock thấp

df["risk_score"] = (
    (
        df["sales_qty"] * 0.4
        + df["forecast_demand"] * 0.6
    )
    / df["current_stock"]
)

# Top 9.25% là cháy hàng

threshold = df["risk_score"].quantile(0.9075)

df["stockout_risk"] = (
    df["risk_score"] >= threshold
).astype(int)

# Remove helper column

df = df.drop(columns=["risk_score"])

# =========================================================
# 9. SORT DATA
# =========================================================

df = df.sort_values(
    by=["date", "store_id", "sku_id"]
).reset_index(drop=True)

# =========================================================
# 10. CHECK RESULT
# =========================================================

print("=" * 60)

print(df.head(10))

print("=" * 60)

print(f"Tổng số dòng dữ liệu: {len(df)}")

stockout_ratio = (
    df["stockout_risk"].sum()
    / len(df)
) * 100

print(
    f"Tỷ lệ Cháy hàng: "
    f"{round(stockout_ratio, 2)}%"
)

print(
    f"Tỷ lệ Đủ hàng: "
    f"{round(100 - stockout_ratio, 2)}%"
)

print("=" * 60)

print(df.describe())

# =========================================================
# 11. EXPORT CSV
# =========================================================

file_name = "TGDĐ_2025_2026_Inventory_Data.csv"

df.to_csv(
    file_name,
    index=False,
    encoding="utf-8-sig"
)

print("=" * 60)

print(f"Đã lưu file: {file_name}")

print("=" * 60)

```

```python
# ==========================================
# SPRINT 1: CÀI ĐẶT MÔI TRƯỜNG & TẢI DỮ LIỆU
# ==========================================
# 1. Cài đặt các thư viện lõi (thêm -q để ẩn bớt log hệ thống)
!pip install -q xgboost lightgbm imbalanced-learn scikit-learn pandas numpy matplotlib seaborn

import os
import pandas as pd
import numpy as np
import warnings

# Tắt cảnh báo để báo cáo đầu ra gọn gàng, chuyên nghiệp
warnings.filterwarnings('ignore')

print("🚀 BẮT ĐẦU SPRINT 1: NẠP VÀ CHUẨN HÓA DỮ LIỆU")
file_name = 'TGDĐ_2025_2026_Inventory_Data (2) (1).csv'

# Kiểm tra xem file đã được tải lên Colab chưa
if not os.path.exists(file_name):
    raise FileNotFoundError(f"❌ LỖI NGHIÊM TRỌNG: Không tìm thấy file '{file_name}'. Vui lòng tải file lên Colab trước khi chạy!")

# 2. Đọc file dữ liệu CSV
print("-> Đang đọc file CSV...")
# Tự động nhận diện dải phân cách (dấu phẩy) từ file dữ liệu mới của bạn
df = pd.read_csv(file_name)

# 3. Chuẩn hóa tên cột theo đúng thiết kế cơ sở dữ liệu
df.columns = ['date', 'store_id', 'sku_id', 'sales_qty', 'lead_time_days',
              'current_stock', 'promo_flag', 'forecast_demand', 'stockout_risk']

# 4. Ép kiểu dữ liệu thời gian (Quan trọng nhất để chạy Time-Series)
print("-> Đang xử lý định dạng thời gian...")
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Bỏ các dòng bị lỗi ngày tháng (nếu có) và sắp xếp tăng dần theo thời gian
df = df.dropna(subset=['date']).sort_values(by='date').reset_index(drop=True)

# 5. In kết quả kiểm tra (Nghiệm thu Sprint 1)
print("\n✅ HOÀN TẤT SPRINT 1!")
print("-" * 50)
print(f"Tổng số dòng dữ liệu hợp lệ: {df.shape[0]:,} dòng")
print(f"Số cột dữ liệu: {df.shape[1]} cột")
print(f"Khung thời gian dữ liệu: Từ {df['date'].dt.date.min()} đến {df['date'].dt.date.max()}")
print("\n[Preview 5 dòng dữ liệu đầu tiên]")
display(df.head())
```

```python
# ==========================================
# SPRINT 2: KỸ NGHỆ ĐẶC TRƯNG & QUICK-WIN REGRESSION
# ==========================================
import numpy as np
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("🚀 BẮT ĐẦU SPRINT 2: TẠO ĐẶC TRƯNG & HUẤN LUYỆN QUICK-WIN")

# 1. Cyclical Encoding (Mã hóa chu kỳ thời gian)
print("-> Đang mã hóa chu kỳ thời gian (Tháng, Ngày trong tuần)...")
df['month'] = df['date'].dt.month
df['day_of_week'] = df['date'].dt.dayofweek

df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

# 2. Lag Features (Tạo biến trễ lịch sử doanh số)
print("-> Đang tạo Lag Features (Doanh số 1, 3, 7 ngày trước)...")
for lag in [1, 3, 7]:
    df[f'sales_qty_lag_{lag}'] = df.groupby(['store_id', 'sku_id'])['sales_qty'].shift(lag)

# Loại bỏ các dòng bị rỗng (NaN) do hàm shift tạo ra
df = df.dropna().reset_index(drop=True)

# 3. Log1p Transformation (Chuẩn hóa Logarit biến mục tiêu)
print("-> Đang chuẩn hóa Logarit cho biến mục tiêu (sales_qty)...")
df['sales_qty_log1p'] = np.log1p(df['sales_qty'])

# Mã hóa thêm các biến Category sang dạng số để đưa vào mô hình
df['store_id_encoded'] = df['store_id'].astype('category').cat.codes.astype(int)
df['sku_id_encoded'] = df['sku_id'].astype('category').cat.codes.astype(int)

# Khai báo danh sách các đặc trưng (features)
features = ['store_id_encoded', 'sku_id_encoded', 'lead_time_days', 'current_stock', 'promo_flag',
            'month_sin', 'month_cos', 'day_sin', 'day_cos',
            'sales_qty_lag_1', 'sales_qty_lag_3', 'sales_qty_lag_7']

# 4. Huấn luyện sớm LightGBM (Quick-Win Regression)
print("-> Đang chạy Quick-Win LightGBM để kiểm chứng tính khả thi...")
# Tạm cắt nhanh dữ liệu (80-20) theo thời gian để chạy thử nghiệm sớm.
# (Việc cắt lớp dữ liệu chuẩn 2025-2026 sẽ được thực hiện chuyên sâu ở Sprint 3)
X_qw = df[features]
y_qw = df['sales_qty_log1p']
X_train_qw, X_test_qw, y_train_qw, y_test_qw = train_test_split(X_qw, y_qw, test_size=0.2, shuffle=False)

# Chạy mô hình
reg_model_qw = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.05, random_state=42)
reg_model_qw.fit(X_train_qw, y_train_qw)

# Đo lường
y_pred_qw_log = reg_model_qw.predict(X_test_qw)
y_pred_qw = np.expm1(y_pred_qw_log) # Nghịch đảo Logarit về số lượng thực
y_test_qw_actual = np.expm1(y_test_qw)

rmse_qw = np.sqrt(mean_squared_error(y_test_qw_actual, y_pred_qw))
mae_qw = mean_absolute_error(y_test_qw_actual, y_pred_qw)

print("\n✅ HOÀN TẤT SPRINT 2!")
print("-" * 50)
print(f"🎯 KẾT QUẢ ĐẦU RA (DELIVERABLE):")
print(f"   - RMSE: {rmse_qw:.2f} sản phẩm")
print(f"   - MAE:  {mae_qw:.2f} sản phẩm")
if mae_qw < 10:
    print("   -> ĐẠT MỤC TIÊU: Sai số MAE dưới 10 sản phẩm/ngày. Dự án hoàn toàn khả thi!")
```

```python
# ==========================================
# SPRINT 3: CÔ LẬP DỮ LIỆU & XỬ LÝ MẤT CÂN BẰNG (SMOTE)
# ==========================================
from imblearn.over_sampling import SMOTE

print("🚀 BẮT ĐẦU SPRINT 3: CÔ LẬP DỮ LIỆU VÀ SMOTE")

# 1. Phân chia dữ liệu theo Khung thời gian (Theo Outline Bảng 3.3)
print("-> Đang phân chia dữ liệu Train/Val/Test theo đúng lịch trình thực tế...")
train_mask = (df['date'] >= '2025-01-01') & (df['date'] <= '2025-12-31')
val_mask = (df['date'] >= '2026-01-01') & (df['date'] <= '2026-03-31')
test_mask = (df['date'] >= '2026-04-01') & (df['date'] <= '2026-05-31')

train_df = df[train_mask]
val_df = df[val_mask]
test_df = df[test_mask]

# Kiểm tra dự phòng nếu tệp CSV bị thiếu dữ liệu tháng 4-5/2026
if test_df.empty:
    print("   ⚠️ Cảnh báo: Dữ liệu Test tháng 04-05/2026 trống, tự động chuyển về chia tỷ lệ 70-15-15.")
    train_size = int(len(df) * 0.7)
    val_size = int(len(df) * 0.15)
    train_df = df.iloc[:train_size]
    val_df = df.iloc[train_size:train_size+val_size]
    test_df = df.iloc[train_size+val_size:]

print(f"   - Tập Huấn luyện (Train): {train_df.shape[0]:,} dòng")
print(f"   - Tập Tinh chỉnh (Val):   {val_df.shape[0]:,} dòng")
print(f"   - Tập Kiểm thử (Test):    {test_df.shape[0]:,} dòng")

# 2. Chuẩn bị tập (X) và (y) cho bài toán Cảnh báo cháy hàng (Classification)
X_train = train_df[features]
y_train_clf = train_df['stockout_risk']

X_val = val_df[features]
y_val_clf = val_df['stockout_risk']

X_test = test_df[features]
y_test_clf = test_df['stockout_risk']

# 3. Áp dụng SMOTE để xử lý mất cân bằng lớp
print("-> Đang áp dụng thuật toán SMOTE (chỉ trên tập Train)...")
try:
    smote = SMOTE(sampling_strategy=0.35, random_state=42)
    X_train_smote, y_train_clf_smote = smote.fit_resample(X_train, y_train_clf)

    # SMOTE sinh ra dữ liệu nội suy nên có thể tạo ra số thập phân cho ID cửa hàng hoặc Cờ khuyến mãi.
    # Ta BẮT BUỘC phải ép kiểu về số nguyên (int) để thuật toán XGBoost không bị lỗi.
    X_train_smote['store_id_encoded'] = X_train_smote['store_id_encoded'].round().astype(int)
    X_train_smote['sku_id_encoded'] = X_train_smote['sku_id_encoded'].round().astype(int)
    X_train_smote['promo_flag'] = X_train_smote['promo_flag'].round().astype(int)

    print(f"   ✅ Đã cân bằng! Số lượng mẫu cháy hàng (Nhãn 1) được tăng từ {sum(y_train_clf):,} lên {sum(y_train_clf_smote):,} mẫu.")
except ValueError as e:
    print(f"   ⚠️ Bỏ qua SMOTE do lượng mẫu thiểu số quá ít. Lỗi: {e}")
    X_train_smote, y_train_clf_smote = X_train, y_train_clf

print("\n✅ HOÀN TẤT SPRINT 3!")
print("-" * 50)
print("Dữ liệu đã được cô lập hoàn hảo, sẵn sàng để bơm vào thuật toán XGBoost!")
```

```python
# ==========================================
# SPRINT 4: HUẤN LUYỆN HỆ THỐNG CẢNH BÁO ĐỨT GÃY (XGBOOST)
# ==========================================
import xgboost as xgb
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

print("🚀 BẮT ĐẦU SPRINT 4: HUẤN LUYỆN MÔ HÌNH PHÂN LOẠI (CLASSIFICATION)")

# 1. Khởi tạo và huấn luyện mô hình XGBoost
print("-> Đang khởi tạo và huấn luyện thuật toán XGBoost Classifier...")
clf_model = xgb.XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    eval_metric='logloss',
    random_state=42
)

# Đưa tập dữ liệu đã tiêm SMOTE vào huấn luyện, dùng tập Val (Q1/2026) để giám sát
clf_model.fit(
    X_train_smote, y_train_clf_smote,
    eval_set=[(X_val, y_val_clf)],
    verbose=False
)

# 2. Đánh giá mô hình trên tập Test (Tháng 04-05/2026)
print("-> Đang kiểm thử tính chính xác trên tập Test (Dữ liệu chưa từng thấy)...")
y_pred_clf = clf_model.predict(X_test)

print("\n✅ HOÀN TẤT SPRINT 4!")
print("=" * 50)
print("🚨 BÁO CÁO NGHIỆM THU - CẢNH BÁO RỦI RO CHÁY HÀNG (XGBOOST):")
print(classification_report(y_test_clf, y_pred_clf, target_names=['An toàn (0)', 'Nguy cơ cháy hàng (1)'], zero_division=0))
print("=" * 50)

# 3. Trực quan hóa Mức độ quan trọng của đặc trưng (Feature Importance)
# Bước này giúp giải thích cho con người hiểu tại sao AI lại quyết định cảnh báo cháy hàng
fig, ax = plt.subplots(figsize=(10, 6))
xgb.plot_importance(clf_model, max_num_features=10, ax=ax,
                    title="Mức độ quan trọng của các biến số (XGBoost Insight)",
                    importance_type='gain', xlabel='Lượng thông tin học được (Gain)')
plt.tight_layout()
plt.show()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import lightgbm as lgb
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')

st.set_page_config(page_title="TGDĐ - AI Inventory Management", page_icon="📱", layout="wide")

# SỬA LỖI CACHE: Dùng cache_data thay vì cache_resource khi trả về DataFrame
@st.cache_data
def run_ai_pipeline():
    # 1. Nạp và chuẩn hóa
    df = pd.read_csv('TGDĐ_2025_2026_Inventory_Data (2) (1).csv', sep=None, engine='python')
    df.columns = ['date', 'store_id', 'sku_id', 'sales_qty', 'lead_time_days',
                  'current_stock', 'promo_flag', 'forecast_demand', 'stockout_risk']
    df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
    df = df.dropna(subset=['date']).sort_values(by='date').reset_index(drop=True)

    # 2. Đặc trưng thời gian & Lag
    df['month'] = df['date'].dt.month
    df['day_of_week'] = df['date'].dt.dayofweek
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df['day_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
    df['day_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
    df['sales_qty_log1p'] = np.log1p(df['sales_qty'])

    for lag in [1, 3, 7]:
        df[f'sales_qty_lag_{lag}'] = df.groupby(['store_id', 'sku_id'])['sales_qty'].shift(lag)

    df = df.dropna().reset_index(drop=True)

    # Ép kiểu int rõ ràng để tránh lỗi XGBoost
    df['store_id_encoded'] = df['store_id'].astype('category').cat.codes.astype(int)
    df['sku_id_encoded'] = df['sku_id'].astype('category').cat.codes.astype(int)
    df['promo_flag'] = df['promo_flag'].astype(int)

    features = ['store_id_encoded', 'sku_id_encoded', 'lead_time_days', 'current_stock', 'promo_flag',
                'month_sin', 'month_cos', 'day_sin', 'day_cos',
                'sales_qty_lag_1', 'sales_qty_lag_3', 'sales_qty_lag_7']

    X = df[features]

    # 3. Chạy thuật toán để lấy kết quả xuất lên Web
    reg_model = lgb.LGBMRegressor(n_estimators=100, learning_rate=0.05, random_state=42)
    reg_model.fit(X, df['sales_qty_log1p'])
    df['AI_Daily_Forecast'] = np.expm1(reg_model.predict(X)).round()

    clf_model = xgb.XGBClassifier(n_estimators=100, learning_rate=0.05, eval_metric='logloss', random_state=42)
    clf_model.fit(X, df['stockout_risk'])
    df['AI_Risk_Warning'] = clf_model.predict(X)

    return df

try:
    df = run_ai_pipeline()
except Exception as e:
    st.error(f"❌ Lỗi tải dữ liệu hoặc huấn luyện AI: {e}")
    st.stop()

# ==========================================
# GIAO DIỆN CHÍNH
# ==========================================
st.sidebar.image("https://haitrieu.com/blogs/vector-logo-the-gioi-di-dong-mwg/?srsltid=AfmBOoq9hpD3JwN9eW029CGjnXdFr_yasEUdCLPRUnsA31OZgQnWGKTZ", width=200)
st.sidebar.markdown("### ⚙️ BỘ LỌC DỮ LIỆU")

store_list = df['store_id'].unique().tolist()
selected_store = st.sidebar.selectbox("🏠 Chọn Siêu thị", ["Tất cả"] + store_list)

sku_list = df['sku_id'].unique().tolist()
selected_sku = st.sidebar.selectbox("📦 Chọn Mã sản phẩm", ["Tất cả"] + sku_list)

filtered_df = df.copy()
if selected_store != "Tất cả":
    filtered_df = filtered_df[filtered_df['store_id'] == selected_store]
if selected_sku != "Tất cả":
    filtered_df = filtered_df[filtered_df['sku_id'] == selected_sku]

st.title("📊 Hệ Thống AI Dự Báo & Tối Ưu Tồn Kho TGDĐ")
st.markdown("---")

col1, col2, col3, col4 = st.columns(4)
total_sales = filtered_df['sales_qty'].sum()
avg_lead_time = filtered_df['lead_time_days'].mean()
total_stockouts = filtered_df[filtered_df['AI_Risk_Warning'] == 1].shape[0]
avg_stock = filtered_df['current_stock'].mean()

col1.metric("Tổng lượng bán", f"{total_sales:,.0f} máy")
col2.metric("Tồn kho trung bình", f"{avg_stock:,.0f} máy")
col3.metric("Lead Time trung bình", f"{avg_lead_time:.1f} ngày")
col4.metric("🚨 Cảnh báo đứt gãy", f"{total_stockouts} lượt")

st.markdown("---")

tab1, tab2, tab3 = st.tabs(["📉 Tổng quan Dự báo", "🚨 Cảnh báo Rủi ro", "📦 Khuyến nghị Nhập hàng"])

with tab1:
    st.subheader("Biểu đồ So sánh Nhu cầu: Thực tế vs AI Dự báo")
    daily_trend = filtered_df.groupby('date')[['sales_qty', 'AI_Daily_Forecast']].sum().reset_index()
    fig, ax = plt.subplots(figsize=(12, 5))
    ax.plot(daily_trend['date'], daily_trend['sales_qty'], label='Thực tế bán ra', color='blue', marker='o', alpha=0.6)
    ax.plot(daily_trend['date'], daily_trend['AI_Daily_Forecast'], label='AI Dự báo (LightGBM)', color='orange', linestyle='--', linewidth=2)
    ax.legend()
    ax.grid(True, linestyle=':', alpha=0.6)
    st.pyplot(fig)

with tab2:
    st.subheader("Danh sách SKU Rủi ro Cháy Hàng")
    risk_df = filtered_df[filtered_df['AI_Risk_Warning'] == 1]
    if risk_df.empty:
        st.success("✅ Hiện tại không phát hiện rủi ro đứt gãy cung ứng.")
    else:
        st.error(f"⚠️ Phát hiện {len(risk_df)} điểm nghẽn cung ứng!")
        st.dataframe(risk_df[['date', 'store_id', 'sku_id', 'sales_qty', 'current_stock', 'lead_time_days']].sort_values(by='date', ascending=False), use_container_width=True)

with tab3:
    st.subheader("Hệ thống Tự động Đề xuất Nhập hàng")
    suggest_df = filtered_df.groupby(['store_id', 'sku_id']).agg({
        'current_stock': 'last',
        'AI_Daily_Forecast': 'mean',
        'lead_time_days': 'mean'
    }).reset_index()

    suggest_df['forecast_demand'] = np.ceil(suggest_df['AI_Daily_Forecast'] * suggest_df['lead_time_days'])
    suggest_df['Safety_Stock'] = np.ceil(suggest_df['forecast_demand'] * 0.5)
    suggest_df['Reorder_Qty'] = np.maximum(0, suggest_df['forecast_demand'] + suggest_df['Safety_Stock'] - suggest_df['current_stock'])
    action_df = suggest_df[suggest_df['Reorder_Qty'] > 0].sort_values(by='Reorder_Qty', ascending=False)

    if action_df.empty:
        st.success("Tất cả hàng hóa đang ở mức an toàn.")
    else:
        st.dataframe(action_df[['store_id', 'sku_id', 'forecast_demand', 'Safety_Stock', 'current_stock', 'Reorder_Qty']].style.highlight_max(subset=['Reorder_Qty'], color='lightcoral'), use_container_width=True)
```
